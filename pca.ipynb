```python
# ==============================
# IMPORT LIBRARIES
# ==============================

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# ==============================
# LOAD DATASET
# ==============================

# The previous error indicated that 'spotify_tracks.csv' was not found at the specified path.
# Please ensure the file is uploaded to '/content/drive/MyDrive/' or update the path below.
# For demonstration, I will try to use a placeholder or ask for user input to fix this.
# If you have the file in a different location, please update the path.
# For now, I will comment out the line causing the error and add a placeholder.

# df = pd.read_csv('/content/drive/MyDrive/spotify_tracks.csv')

# To proceed, you would need to either:
# 1. Upload 'spotify_tracks.csv' to your Google Drive and ensure it's in the specified path.
# 2. Modify the path to where your 'spotify_tracks.csv' is located.
# 3. If you want to use a sample dataset for testing, uncomment the line below:
# df = pd.DataFrame({
#     'danceability': [0.8, 0.7, 0.6, 0.9, 0.5],
#     'energy': [0.9, 0.6, 0.7, 0.8, 0.4],
#     'loudness': [-3, -7, -5, -4, -10],
#     ''tempo': [120, 100, 130, 110, 90],
#     'acousticness': [0.1, 0.5, 0.2, 0.05, 0.8],
#     'valence': [0.7, 0.4, 0.6, 0.9, 0.3],
#     'speechiness': [0.05, 0.1, 0.03, 0.08, 0.15]
# })

# As a temporary fix to allow the code to run without the file,
# I'll create a dummy DataFrame. You should replace this with your actual data loading.
df = pd.DataFrame({
    'danceability': [0.8, 0.7, 0.6, 0.9, 0.5, 0.75, 0.65, 0.85, 0.55, 0.7],
    'energy': [0.9, 0.6, 0.7, 0.8, 0.4, 0.8, 0.68, 0.88, 0.48, 0.75],
    'loudness': [-3, -7, -5, -4, -10, -3.5, -6.5, -4.5, -9.5, -5.5],
    'tempo': [120, 100, 130, 110, 90, 125, 105, 135, 95, 115],
    'acousticness': [0.1, 0.5, 0.2, 0.05, 0.8, 0.15, 0.45, 0.1, 0.75, 0.3],
    'valence': [0.7, 0.4, 0.6, 0.9, 0.3, 0.75, 0.45, 0.95, 0.35, 0.65],
    'speechiness': [0.05, 0.1, 0.03, 0.08, 0.15, 0.06, 0.12, 0.04, 0.18, 0.09]
})

# ==============================
# SELECT FEATURES
# ==============================

features = [
    'danceability',
    'energy',
    'loudness',
    'tempo',
    'acousticness',
    'valence',
    'speechiness'
]

X = df[features]

# ==============================
# STANDARDIZE DATA
# ==============================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ==============================
# APPLY PCA
# ==============================

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

# ==============================
# CREATE NEW DATAFRAME
# ==============================

pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# ==============================
# PLOT MUSIC SPACE
# ==============================

plt.figure(figsize=(8,6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], alpha=0.5)

plt.title("🎵 2D Music Space Visualization (PCA)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.show()
```
