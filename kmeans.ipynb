```python
# =====================================
# INSTALL LIBRARIES
# =====================================
!pip install gradio scikit-learn pandas numpy matplotlib

# =====================================
# IMPORT LIBRARIES
# =====================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gradio as gr

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# =====================================
# LOAD DATASET FROM DRIVE
# =====================================
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/spotify_tracks.csv')

df = df.dropna()

# =====================================
# SELECT FEATURES FOR CLUSTERING
# =====================================
features = [
    'danceability',
    'energy',
    'tempo',
    'valence',
    'acousticness'
]

X = df[features]

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =====================================
# APPLY K-MEANS
# =====================================
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

print(df['Cluster'].value_counts())

# =====================================
# MANUAL CLUSTER LABELING
# =====================================
cluster_labels = {
    0: "🎧 Chill Study",
    1: "💪 Workout",
    2: "💖 Romantic",
    3: "🎉 Party"
}

df['Playlist_Type'] = df['Cluster'].map(cluster_labels)

# =====================================
# CREATIVE UI FUNCTION
# =====================================
def recommend_playlist(danceability, energy, tempo, valence, acousticness):

    input_data = np.array([[danceability, energy, tempo, valence, acousticness]])
    input_scaled = scaler.transform(input_data)

    cluster = kmeans.predict(input_scaled)[0]
    playlist = cluster_labels[cluster]

    # Create cluster probability distance chart
    distances = kmeans.transform(input_scaled)[0]

    chart_data = pd.DataFrame({
        "Cluster": ["Chill", "Workout", "Romantic", "Party"],
        "Distance Score": distances
    })

    return playlist, chart_data


# =====================================
# CREATIVE UI WITH BARCHART
# =====================================
with gr.Blocks() as app:

    gr.Markdown("# 🎵 AI Automatic Playlist Generator")
    gr.Markdown("### Discover your vibe instantly!")

    with gr.Row():
        with gr.Column():
            danceability = gr.Slider(0,1,label="💃 Danceability")
            energy = gr.Slider(0,1,label="⚡ Energy")
            tempo = gr.Slider(0,250,label="🥁 Tempo")
            valence = gr.Slider(0,1,label="😊 Valence (Mood)")
            acousticness = gr.Slider(0,1,label="🎸 Acousticness")

            btn = gr.Button("✨ Generate Playlist")

        with gr.Column():
            result = gr.Textbox(label="Recommended Playlist")
            chart = gr.BarPlot(
                x="Cluster",
                y="Distance Score",
                title="Similarity to Each Playlist"
            )

    btn.click(
        recommend_playlist,
        inputs=[danceability, energy, tempo, valence, acousticness],
        outputs=[result, chart]
    )

app.launch(share=True)
```
